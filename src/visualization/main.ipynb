```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DATASET_DIR = "../../raw-dataset/DAIC-WOZ-description/"
```

```python
all_data = pd.read_csv(DATASET_DIR + "detailed_lables.csv")

train_data = pd.read_csv(DATASET_DIR + "train_split.csv")
dev_data = pd.read_csv(DATASET_DIR + "dev_split.csv")
test_data = pd.read_csv(DATASET_DIR + "test_split.csv")

print("All data shape: ", all_data.shape)
print("Train data shape: ", train_data.shape)
print("Dev data shape: ", dev_data.shape)
print("Test data shape: ", test_data.shape)
```

```python
# depressed vs not depressed distribution in train and dev data
train_depressed = train_data[train_data["PHQ_Binary"] == 1]
dev_depressed = dev_data[dev_data["PHQ_Binary"] == 1]

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.countplot(x="PHQ_Binary", data=train_data, hue="PHQ_Binary")
plt.title("Train data distribution")
plt.subplot(1, 2, 2)
sns.countplot(x="PHQ_Binary", data=dev_data, hue="PHQ_Binary")
plt.title("Dev data distribution")
plt.show()
```

```python
# gender distribution in train and dev data
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.countplot(x="Gender", data=train_data, hue="Gender")
plt.title("Gender Distribution in train data")
plt.subplot(1, 2, 2)
sns.countplot(x="Gender", data=dev_data, hue="Gender")
plt.title("Gender Distribution in dev data distribution")
plt.show()
```

```python
# plot of gender distribution in depressed and not depressed patients
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.countplot(x="Gender", data=train_data, hue="PHQ_Binary")
plt.title("Gender and PHQ_Binary distribution in train data")
plt.subplot(1, 2, 2)
sns.countplot(x="Gender", data=dev_data, hue="PHQ_Binary")
plt.title("Gender and PHQ_Binary distribution in test data")
plt.show()
```
